## Sección Técnica Entrevista Quantsoft
Modelaremos el comportamiento de la acciones de Microsoft. Para ello, utilizamos la base de datos disponible desde la API de Yahoo Finance

```python
import yfinance as yf
import matplotlib.pyplot as plt
msft = yf.Ticker("MSFT")
```

Explorammos el dataset de acciones de MFST durante un intervalo de tiempo específico, desde 2010-2022

```python
# stock info
df = msft.history(period='1mo',start='2010-01-01',end='2022-12-31')
print(df.head())
```

Para comenzar, visualizaremos una serie temporal sobre este período. La visualización ayuda rápidamente a detectar cómo el valor asciende rápidamente a partir del año 2015-2016, momento en que supera el threshold previo de 50. El valor vuelve a aumentar rápidamente a partir del año 2020

```python
fig = plt.figure(figsize=[10.5,4.5])
ax = fig.add_subplot(111)
ax.plot(df.index,df.Close)
ax.set_xlabel('Time')
ax.set_ylabel('Close value')
```

Primero, observamos el inicio de la curva, antes del año 2018. En un inicio, la curva se mantiene sin grandes variaciones durante el año 2013

```python
df_2018 = df[df.index<='2018-01-01']
fig = plt.figure(figsize=[10.5,4.5])
ax = fig.add_subplot(111)
ax.plot(df_2018.index,df_2018.Close)
```

Podemos observar también la distribución del valor de la acción. Durante este período, predominan valores sub 50. Los altos valores de la cola no se condicen con el comportamiento que se observaba hasta la fecha.

```python
fig = plt.figure(figsize=[10.5,4.5])
ax = fig.add_subplot(111)
ax.hist(df_2018.Close,density=False,bins=20)
ax.set_xlabel('Close value')
ax.set_ylabel('Count')
```

Hacemos un zoom para visualizar mejor los valores a partir del año 2018. Enero de 2021 parece no sigue la tendencia de un aumento relativo respecto a meses anteriores, el cual es mucho más evidente en el año 2022. Sin embargo, parece sí haber un aumento relativo durante los meses de septiembre.

```python
df_2020 = df[df.index>='2018-01-01']
fig = plt.figure(figsize=[10.5,4.5])
ax = fig.add_subplot(111)
ax.plot(df_2020.index,df_2020.Close)
```

Comenzamos con una comparativa rápida del valor promedio en cada año. Recién tras 5 años se observa una variación importante, a medida que el promedio asciende rápidamente.

```python
df.Close.groupby(df.index.year).mean()
```

Visualizamos el cambio porcentual entre el valor en un instante respecto al valor inmediatamente anterior de la acción a lo largo de este período. El valor de la acción se mantiene dentro de rangos inferiores a 5%. Precisamente en el período final del año 2019 las variaciones son mínimas, para luego cambiar abruptamente una vez inicia el año 2020. Los primeros cambios corresponden a variaciones fuertemente negativas

```python
fig = plt.figure(figsize=[10.5,4.5])
ax = fig.add_subplot(111)
ax.plot(df.index,df.Close.pct_change())
ax.set_xlabel('Year')
ax.set_ylabel('Percentual change')
```
